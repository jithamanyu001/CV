```python
import matplotlib.pyplot as plt

import keras_ocr

# keras-ocr will automatically download pretrained
# weights for the detector and recognizer.
pipeline = keras_ocr.pipeline.Pipeline()

# Get a set of three example images
images = [
    keras_ocr.tools.read(img) for img in [
        "/content/img.jpg"
    ]
]

# Each list of predictions in prediction_groups is a list of
# (word, box) tuples.
prediction_groups = pipeline.recognize(images)
```

```python
prediction_groups[0]
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
keras_ocr.tools.drawAnnotations(image=images[0], predictions=prediction_groups[0],ax=axs)
```

```python
import easyocr
reader = easyocr.Reader(['en']) # this needs to run only once to load the model into memory
```

```python
result
```

```python
import cv2
c_img=cv2.imread("/content/img.jpg")
from google.colab.patches import cv2_imshow

for i, r in enumerate(result):
  x_min, y_min = r[0][0]
  x_max, y_max = r[0][2]
  img = cv2.rectangle(c_img, (int(x_min), int(y_min)), (int(x_max), int(y_max)), color = (255, 0,0), thickness=2)
  img = cv2.putText(img, r[1] + " " + f"{r[2]:0.2f}", (int(x_max), int(y_max+10)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,0,0), 1)
cv2_imshow(c_img)
```

```python
import cv2
import pandas as pd
import os
import numpy as np
import os
```

```python
c_img=cv2.imread("/content/img.jpg")
```

```python
from paddleocr import PaddleOCR,draw_ocr
ocr = PaddleOCR(lang='en') # need to run only once to download and load model into memory

result = ocr.ocr(c_img)
#for line in result:
    #print(line)

# draw result

boxes = [line[0] for line in result[0]]
txts = [line[1][0] for line in result[0]]
scores = [line[1][1] for line in result[0]]
print(len(boxes[0]))

for i, box in enumerate(boxes):
  x_min, y_min = box[0]
  x_max, y_max = box[2]
  img = cv2.rectangle(c_img, (int(x_min), int(y_min)), (int(x_max), int(y_max)), color = (255, 0, 0), thickness=2)
  img = cv2.putText(img, txts[i] + " " + f"{scores[i]:0.2f}", (int(x_max), int(y_max+10)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (36,255,12), 1)
cv2_imshow(c_img)
```
